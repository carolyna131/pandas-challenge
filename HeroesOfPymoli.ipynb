```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# CSV File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read CSV Purchasing File and store into dataframe
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(20)
```

## Player Count

* Display the total number of players


```python
#Get number of unique players by removing dups
unique_players = purchase_data.drop_duplicates(subset=["SN"])

#array containing the number of unique players by SN
unique_SN = len(unique_players['SN'].value_counts()) 
total_players_df = pd.DataFrame({"Total Players": [unique_SN]})

#display results
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculations for unique itmes, average price,number of purchases, and total revenue
count_unique_items = len(purchase_data['Item ID'].unique())
count_of_purchases = len(purchase_data['Purchase ID'].unique())
price_sum ='${:,.2f}'.format(sum(purchase_data['Price']))
avg_item_price = '${:,.2f}'.format(sum(purchase_data['Price'])/len(purchase_data['Purchase ID']))

#create dataframe with purchase analysis calculation results
purchase_analysis_df = pd.DataFrame({'Number of Unique Items': [count_unique_items],
                                    'Average Price': [avg_item_price],
                                    'Number of Purchases' : [count_of_purchases],
                                    'Total Revenues': [price_sum]
                                     })
#display results
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Get number of unique players by gender (removing dups)
gender_df = purchase_data[["Gender","SN"]].drop_duplicates(subset ="SN")

#Calculate the counts and precent by gender
gender_count = gender_df["Gender"].value_counts(0)
gender_percent = gender_df["Gender"].value_counts(1)

#create dataframe to populate count and percent
gender_count_df = pd.DataFrame(gender_count)
gender_percent_df = pd.DataFrame(round(gender_percent * 100,2))

#merge the gender count and percent dataframes
gender_summary_df = gender_count_df.merge(gender_percent_df,left_index =True, right_index = True)
gender_summary_df.columns = ["Total Count", "Percentage of Players"]

#display and format results
gender_summary_df.style.format({"Percentage of Players":'{:.2f}%'})
```

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create age bins and labels
age_bins = [0, 9.90, 14.90, 19.90, 24.9, 29.9, 34.90, 39.90, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

#Use Cut to categorize players
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

# count purchases by age bin by counting unique screen names
purchase_count_age = pd.DataFrame(purchase_data.groupby("Age Ranges")['SN'].count())

#find avg price of purchases by age bin
avg_price_age = pd.DataFrame(purchase_data.groupby("Age Ranges")['Price'].mean())

#find total purchase value by age bin
total_purchase_age = pd.DataFrame(purchase_data.groupby("Age Ranges")['Price'].sum())

#delete multiple occurances of SN while only keeping last, then count # of unique
no_dup_age = pd.DataFrame(purchase_data.drop_duplicates('SN', keep = 'last').groupby("Age Ranges")['SN'].count())

#merge all info from above into one df
merge_age = pd.merge(purchase_count_age, avg_price_age, left_index = True, right_index = True).merge(total_purchase_age, left_index = True, right_index = True).merge(no_dup_age, left_index = True, right_index = True)

#rename columns
merge_age.rename(columns = {"SN_x": "Purchase Count", "Price_x": "Average Purchase Price", "Price_y": "Total Purchase Value","SN_y": "# of Purchasers"}, inplace = True)

#calculate avg total purchase by person within the age bin
merge_age['Avg Total Purchase per Person'] = merge_age['Total Purchase Value']/merge_age['# of Purchasers']

#format monetary fields correctly
merge_age.style.format({'Average Purchase Price': '${:.2f}', 'Total Purchase Value': '${:.2f}','Avg Total Purchase per Person': '${:.2f}'})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculate the purchase counts, avg purchase price, total purchase value and avg total per person
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
user_average = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

#create dataframe to populate results
user_data = pd.DataFrame({"Purchase Count": user_count,
                          "Average Purchase Price": user_average,
                          "Total Purchase Value": user_total,
                          })

#display results
user_data.sort_values("Total Purchase Value", ascending=False, inplace=True)

# take top 5 only
user_data = user_data.head(5)

# format
user_data.style.format({'Total Purchase Value': '${:.2f}', 'Average Purchase Price': '${:.2f}'})

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

most_popular_items = purchase_data.groupby(["Item ID","Item Name"])
most_popular_totals = most_popular_items["Item ID", "Item Name"].count()

#Calculate the purchase counts, avg purchase price, total purchase value and avg total per person
most_popular_count = purchase_data.groupby(["Item ID","Item Name"]).count()["Price"].rename("Purchase Count")
most_popular_average = purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Average Purchase Price")
most_popular_total = purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")

#create dataframe to populate results
most_popular_data = pd.DataFrame({"Purchase Count": most_popular_count,
                        "Average Purchase Price": most_popular_average,
                         "Total Purchase Value": most_popular_total,
                        })

#sort and display results
most_popular_data.sort_values("Purchase Count", ascending = False, inplace=True)

# take top 5 only
most_popular_data = most_popular_data.head(5)

# format
most_popular_data.style.format({'Average Purchase Price': '${:.2f}', 'Total Purchase Value': '${:.2f}'})
```
